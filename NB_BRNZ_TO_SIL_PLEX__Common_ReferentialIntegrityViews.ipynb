# Referential Integrity Statistics
<br>

The audit_referentialintegrity view tracks the referential integrity between <BR>
related tables in the model by counting the number of nulls as well as the hash <BR>
values of nulls, empty strings, and zeros stored in the foreign keys of the BaseTable.  <BR>
It also counts the number of missing and valid foreign records in the ForeignTable. <BR>
The view provides insights into the quality of data relationships between tables <BR>
and can be used to identify and resolve data inconsistencies.
