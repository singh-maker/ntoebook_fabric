```python
def generate_merge_sql(database_name, table_name, non_audit_columns, primary_keys, source_table):
    # Extract column names from dictionary
    non_audit_col_names = [col["col_name"] for col in non_audit_columns]
    
    # Create a mapping of column names to data types
    data_type_map = {col["col_name"]: col["data_type"] for col in non_audit_columns}
    
    # Generate primary key hash expression (with backticks inside CAST)
    pk_hash_expr = ",\n              ".join(
        [f"IFNULL(CAST(`{pk}` AS STRING),'NULL')" for pk in primary_keys]
    )
    
    # Generate full row hash expression (with backticks inside CAST)
    row_hash_expr = ",\n              ".join(
        [f"IFNULL(CAST(`{col}` AS STRING),'NULL')" for col in non_audit_col_names]
    )
    
    # Generate column list for SRC1 with explicit casting and aliases
    src1_columns = []
    for col in non_audit_columns:
        col_name = col["col_name"]
        data_type = col["data_type"]
        # Create alias name by replacing special characters and making it camelCase
        alias_name = ''.join(word.capitalize() for word in col_name.lower().split('_'))
        src1_columns.append(f"CAST(`{col_name}` AS {data_type}) AS {alias_name}")
    
    src1_columns_str = ",\n        ".join(src1_columns)
    
    # Generate column list for SRC2 (using aliases from SRC1)
    src2_columns = ",\n     ".join(
        [''.join(word.capitalize() for word in col["col_name"].lower().split('_')) 
         for col in non_audit_columns]
    )
    
    # Generate update set clause (using original column names)
    update_set = ",\n   ".join([f"DST.{col} = SRC.{col}" for col in non_audit_col_names])
    
    # Generate insert columns and values (using original column names)
    insert_cols = ",\n     ".join(non_audit_col_names)
    insert_vals = ",\n     ".join([f"SRC.{col}" for col in non_audit_col_names])
    
    # Generate merge condition (without quotes)
    merge_condition = "\n  AND ".join([f"SRC.{pk}<=>DST.{pk}" for pk in primary_keys])
    
    # Generate partition by clause (without quotes)
    partition_by = ", ".join(primary_keys)
    
    return f"""
_bionSqlResults = spark.sql(\"\"\"
MERGE INTO {database_name}.{table_name} DST
USING (
  SELECT 
   unhex(SHA2(concat_ws('|',
              {row_hash_expr}
             ),256)) AS RevisionHash
   , *
  FROM (
   SELECT 
     unhex({table_name}Key__HEXSTRING) AS {table_name}Key,
     cast(conv(left({table_name}Key__HEXSTRING,15),16,10) as bigint) AS {table_name}KeyInt,
     {src2_columns},
     Audit_Source_ModifiedDateTime AS Audit_Source_ModifiedDateTime,
     ROW_NUMBER() OVER (PARTITION BY {partition_by} ORDER BY Audit_Source_ModifiedDateTime DESC) DUPLICATES_RN,
     RIGHTNOW AS RIGHTNOW          
   FROM (
     SELECT
        SHA2(concat_ws('|',
              {pk_hash_expr}
             ),256) AS {table_name}Key__HEXSTRING,
        {src1_columns_str},
        Audit_Source_ModifiedDateTime AS Audit_Source_ModifiedDateTime,,
        RIGHTNOW
     FROM {source_table}
     ) SRC1
  ) SRC2 WHERE DUPLICATES_RN = 1
 ) SRC
ON 1=1
AND {merge_condition}
WHEN MATCHED AND SRC.RevisionHash<>IFNULL(DST.RevisionHash,unhex('0000000')) THEN UPDATE SET
   DST.{table_name}Key = SRC.{table_name}Key,
   DST.{table_name}KeyInt = SRC.{table_name}KeyInt,
   {update_set},
   DST.Audit_Source_ModifiedDateTime = SRC.Audit_Source_ModifiedDateTime,
   DST.RevisionHash = SRC.RevisionHash,
   DST.Audit_ModifiedDateTime = SRC.RIGHTNOW
WHEN NOT MATCHED THEN INSERT (
     {table_name}Key,
     {table_name}KeyInt,
     {insert_cols},
     RevisionHash,
     Audit_Source_ModifiedDateTime,
     Audit_CreatedDateTime,
     Audit_ModifiedDateTime
  )
  VALUES(
    SRC.{table_name}Key,
    SRC.{table_name}KeyInt,
    {insert_vals},
    SRC.RevisionHash,
    SRC.Audit_Source_ModifiedDateTime,
    SRC.RIGHTNOW,
    SRC.RIGHTNOW
  )
\"\"\")
"""

# Example usage
database_name = "LH_BI_SILVER_TIER2"
table_name = "cost_model"
primary_keys = ["PCN", "Cost_Model_Key"]
source_table = "src_data_for_cost_model"

non_audit_columns = [
    {"col_name": "PCN", "data_type": "int"},
    {"col_name": "Cost_Model_Key", "data_type": "int"},
    {"col_name": "Cost_Model", "data_type": "string"},
    {"col_name": "Description", "data_type": "string"},
    {"col_name": "Active", "data_type": "string"},
    {"col_name": "Primary_model", "data_type": "string"},
    {"col_name": "Frozen", "data_type": "string"}
]

generated_code = generate_merge_sql(
    database_name,
    table_name,
    non_audit_columns,
    primary_keys,
    source_table
)

print(generated_code)
```
