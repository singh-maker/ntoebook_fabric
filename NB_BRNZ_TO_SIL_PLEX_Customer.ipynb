#### Manage data quality violations marked for auditing

```python
%%sql
--CREATE THE LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations TABLE
CREATE TABLE IF NOT EXISTS LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations
  ( --need at least a basic table to auto merge the schema.
    PCN INT COMMENT '[ID:{29615}]',--Primary Key
    Customer_No STRING COMMENT '[ID:{73045}]',--Primary Key
    Audit_ModifiedDateTime TIMESTAMP
  )
  USING DELTA
```

```python
%%sql
--To fully leverage the schmema merge functionality, we need to first create these views.
--this one identifies fixed records...
CREATE OR REPLACE TEMPORARY VIEW customer_Audit_DataQualityViolations_Fixed AS
  SELECT * from src_data_for_customer
  WHERE (--if we meet all the following conditions then the record has been fixed...
    (PCN is not null)
  )
```

```python
%%sql
--This view identifies records in violation of at least one rule, and adds the "dynamic" Audit columns
CREATE OR REPLACE TEMPORARY VIEW customer_Audit_DataQualityViolations_Current AS
  SELECT 
    *
    ,IFNULL(NOT (PCN is not null),FALSE) AS Audit_IsViolationOf_PCNnotNull
  FROM src_data_for_customer
  WHERE NOT (--if we meet all the following conditions then the record has been fixed...
    (PCN is not null)
  )
```

```python
%%sql
--This section deletes fixed records from the LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations table.
--We always merge schemas on the data quality audit tables so we capture new fields but don't lose old ones...
set spark.databricks.delta.schema.autoMerge.enabled = True;
--We've split the delete, insert and update to accommodate the createddate and other possible future differences.

MERGE INTO LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations DST
USING (
 SELECT * FROM customer_Audit_DataQualityViolations_Fixed
) SRC
ON 
1=1
 AND SRC.PCN<=>DST.PCN
 AND SRC.customer_No<=>DST.customer_No
WHEN MATCHED THEN DELETE --delete (it's fixed now)
```

```python
%%sql
--This section inserts new records that have DQ Violations into the LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations table.
--We always merge schemas on the data quality audit tables so we capture new fields but don't lose old ones...
set spark.databricks.delta.schema.autoMerge.enabled = True;
--We've split the delete, insert and update to accommodate the createddate and other possible future differences.

MERGE INTO LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations DST
USING (
 SELECT * FROM customer_Audit_DataQualityViolations_Current
) SRC
ON 
1=1
 AND SRC.PCN<=>DST.PCN
 AND SRC.customer_No<=>DST.customer_No
WHEN MATCHED THEN UPDATE SET *
```

```python
%%sql
--This section inserts new records that have DQ Violations into the LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations table.
--We always merge schemas on the data quality audit tables so we capture new fields but don't lose old ones...
set spark.databricks.delta.schema.autoMerge.enabled = True;
--We've split the delete, insert and update to accommodate the createddate and other possible future differences.

MERGE INTO LH_BI_SILVER_TIER1.customer_Audit_DataQualityViolations DST
USING (
 SELECT * FROM customer_Audit_DataQualityViolations_Current
) SRC
ON 
1=1
 AND SRC.PCN<=>DST.PCN
 AND SRC.customer_No<=>DST.customer_No
WHEN NOT MATCHED THEN INSERT * --only insert if it does not exist. We will update existing records below
```
